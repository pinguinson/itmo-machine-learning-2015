```python
# Load data

ROOT = r'D:\Projects\Kaggle\OttoGroup'
DATA = ROOT + "/data"

data_train = pd.read_csv(DATA + "\\" + "train.csv")
data_train = data_train.drop('id', axis=1)

data_test = pd.read_csv(DATA + "\\" + "test.csv")
data_test_id = data_test['id']
data_test = data_test.drop('id', axis=1)

data_train.target = data_train.target.apply(lambda x: int(x.split("_")[1]))
```

```python
# train & test
train_target = data_train.target
train = data_train.drop('target', axis=1)

X_train, X_test, y_train, y_test = sk.cross_validation.train_test_split(train, train_target, test_size=0.2, random_state=23)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
color_wheel = {1: "#ccffcc", 2: "#ccffff", 3: "#dfc137", 4: "#59983b",
               5: "#ff5d00", 6: "#0000ff", 7: "#292e37", 8: "#983b59",
               9: "#ffff00"}

colors = pd.Series(index=data_train.index)
for color_key in color_wheel.keys():
    colors[data_train['target'] == color_key] = color_wheel[color_key]
pd.scatter_matrix(data_train[['feat_1', 'feat_2', 'feat_3', 'feat_4']], color=colors, alpha=0.6, figsize=(15, 15), diagonal='hist')
```

```python
fig, axes = plt.subplots(ncols=1, nrows=9)

fig.set_size_inches((12,10))
for target_class in range(1,10):
    data = train[train_target == target_class]
    acc_ser = pd.Series(index=range(1, len(data.columns)+1))
    for ind, col in enumerate(data.columns):
        acc_ser[ind] = data[col].mean()
        
    axes[target_class-1].plot(acc_ser.index, acc_ser)
    axes[target_class-1].set_title("class" + str(target_class))
plt.tight_layout()
```
